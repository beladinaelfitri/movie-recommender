```python
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

import numpy as np
import pandas as pd
```

```python
drivePrefix = '/content/gdrive/My Drive/1301174046_MF_MovieLens/'

driveSuffix = 'ml-latest-small/k2_user.csv'
header = [1, 2]
ulf2 = pd.read_csv(drivePrefix + driveSuffix, sep=',', names =header)

driveSuffix = 'ml-latest-small/k2_item.csv'
ilf2 = pd.read_csv(drivePrefix + driveSuffix, sep=',', names =header)

driveSuffix = 'ml-latest-small/k5_user.csv'
header = [1, 2, 3, 4, 5]
ulf5 = pd.read_csv(drivePrefix + driveSuffix, sep=',', names =header)

driveSuffix = 'ml-latest-small/k5_item.csv'
ilf5 = pd.read_csv(drivePrefix + driveSuffix, sep=',', names =header)
```

```python
ulf2 = ulf2.to_numpy()
ilf2 = ilf2.to_numpy()

ulf5 = ulf5.to_numpy()
ilf5 = ilf5.to_numpy()
```

```python
D = np.dot(ulf2, ilf2.T)
L = np.dot(ulf5, ilf5.T)
```

```python
movie_rec2 = []
for i in range(len(D)):
  movie_rec2.append([np.argsort(-1*D[i])[:10]])
movie_rec2
```

```python
movie_rec5 = []
for i in range(len(L)):
  movie_rec5.append([np.argsort(-1*L[i])[:10]])
movie_rec5
```
